## Assignment to apply knowledge of web scraping in conjunction with Beautiful soup

import necessary libraries

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

specify url to scrap data from

```python
url='https://bwfbadminton.com/players/?char=all&country=TPE&page_size=5000&page_no=1'
```

download contents of page specified by url using requests library

```python
page_contents = requests. get(url)
```

create soup using page contents

```python
soup=BeautifulSoup(page_contents.text)
```

find all links in current page

```python
players_tag=(soup.find_all(attrs={"class": "content"})[1])
```

```python
type(players_tag)
```

```python
players_tag.find_all('div')
```

```python
len(players_tag)
```

```python
df=pd.DataFrame(columns=['name', 'url'])
```

```python
for elem in players_tag:
    if len(elem)==1:
        continue
#   var_elem=elem
    v_name=elem.find('div',{'class','name'}).a['title']
    v_url=elem.find('div',{'class','name'}).a['href']
    
    
    df=df.append({'name':v_name, 'url':v_url}, ignore_index=True)
```
